```
school=open("school_prompt.txt","r")
lines=school.readlines()
p_words=[]
for line in lines[0:]:
    val=line.split(" ")
    for wrd in val:
        if 'p' in wrd:
            p_words.append(wrd)
            
for i in range(len(p_words)):
    if '\n' in p_words[i]:
        p_words[i]=p_words[i][:len(p_words[i])-1]
    
print(p_words)
```
